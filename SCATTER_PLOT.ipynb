```python
import matplotlib.pyplot as plt
DAY = [1,2,3,4,5,6,7]
NUMBER = [2,3,1,4,5,3,6]
plt.scatter(DAY,NUMBER,color='r')          ## changed the color
plt.title("scatter plot",fontsize=12)
plt.xlabel("DAY",fontsize=12)           ## labeled
plt.ylabel("NUMBER",fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
DAY = [1,2,3,4,5,6,7]
NUMBER = [2,3,1,4,5,3,6]
colors=['r','y','g','b','r','r','g']            ## for different colors
size=[120,140,130,150,160,170,190]                     ## for dot size
plt.scatter(DAY,NUMBER,c=colors,s=size,alpha=0.7, marker='*',edgecolor='g', linewidth=2)   ## for dot transparency     ## to change the marker  ##to change the edge color
plt.title("scatter plot",fontsize=12)
plt.xlabel("DAY",fontsize=12)
plt.ylabel("NUMBER",fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
DAY = [1,2,3,4,5,6,7]
NUMBER = [2,3,1,4,5,3,6]
number = [2,4,5,1,6,4,3]
colors=[24,45,15,67,56,72,99]
size=[120,140,130,150,160,170,190]
plt.scatter(DAY,NUMBER,c=colors,s=size,  cmap='viridis')          ## changed the color
t=plt.colorbar()
t.set_label("colorbar of viridis",fontsize=12)
plt.scatter(DAY,number,c=colors,s=size, cmap='BrBG')
w=plt.colorbar()
w.set_label("colorbar of BrBG",fontsize=12)
plt.title("scatter plot",fontsize=12)
plt.xlabel("DAY",fontsize=12)           ## labeled
plt.ylabel("NUMBER",fontsize=12)
plt.show()
```

**PROJECT WITH CSV FILE**

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn')
data=pd.read_csv("/content/2019-05-31-data.csv")
view_count=data['view_count']
likes=data['likes']
ratio=data['ratio']
plt.scatter(view_count,likes,c=ratio,cmap='summer',edgecolor='black',linewidth=1,alpha=0.75)

cbar= plt.colorbar()
cbar.set_label('like/dislike ratio')

plt.xscale('log')                ## to scale the data
plt.yscale('log')

plt.title('trending youtube videos')
plt.xlabel('view count')
plt.ylabel('total likes')
plt.tight_layout()
```
